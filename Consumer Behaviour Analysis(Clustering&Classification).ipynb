```python
correlational_matrix=df.corr()
```

```python
threshold=0.5
highly_correlated_features=np.where(np.abs(correlational_matrix)>threshold)
```

```python
#print(highly_correlated_features)
for f1,f2 in zip(highly_correlated_features[0],highly_correlated_features[1]):
  if f1!=f2:
    print(df.columns[f1]," and ",df.columns[f2]," are highly correlated")
```

```python
threshold_=0.0009
low_correlated_features=np.where(np.abs(correlational_matrix)<threshold_)
```

```python
for f1,f2 in zip(low_correlated_features[0],highly_correlated_features[1]):
  if f1!=f2:
    print(df.columns[f1]," and ",df.columns[f2]," are less correlated")
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
sm=RandomOverSampler()
```

```python
x,y=sm.fit_resample(x_updated,y)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
xsc=ms.fit_transform(x)
```

```python
xsc
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(xsc,y,test_size=0.1,random_state=4)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
y_predicted=knn.predict(x_test)
```

```python
y_predicted
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={"n_neighbors":[3,5,7,9],"weights":["uniform","distance"],"algorithm":["auto","ball_tree","kd_tree","brute"]}
clf=GridSearchCV(knn,parameters,cv=5,scoring="accuracy")
clf.fit(x_train,y_train)
```

```python
print(clf.best_params_)
```

```python
knn1=KNeighborsClassifier(algorithm='auto',n_neighbors=9,weights='distance')
knn1.fit(x_train,y_train)
y_pred1=knn1.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```

# SVC

```python
from sklearn.svm import SVC
```

```python
parameter_svc = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf'], 'gamma': [0.1, 1, 10]}
svc = SVC()

grid_search = GridSearchCV(estimator=svc, param_grid=parameter_svc, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)

best_params = grid_search.best_params_
best_score = grid_search.best_score_

best_svc = SVC(**best_params)
best_svc.fit(x_train, y_train)

accuracy = best_svc.score(x_test, y_test)

print("Best Parameters:", best_params)
print("Best Score:", best_score)
print("Test Accuracy:", accuracy)
```

```python
sv=SVC(C=10, gamma =1, kernel= 'rbf')
sv.fit(x_train,y_train)
y_predicted_svc=sv.predict(x_test)
y_predicted_svc
```

```python
print(classification_report(y_test,y_predicted_svc))
```

# Clustering(Unsupervised)

```python
x_c=df_copy.iloc[:,5:21:14] #5th column number imcluded,21st column number excluded, every 14th column picked in [5,21) range.Start index pf columns with 0.
```

```python
print(x_c)
```

```python
le_c=LabelEncoder()
x_c["Browsing_Frequency"]=le_c.fit_transform(x_c["Browsing_Frequency"])
```

```python
from sklearn.cluster import KMeans
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
wcss=[]
for i in range(1,7):
  km=KMeans(n_clusters=i,init="k-means++",random_state=5)
  km.fit(x_c)
  wcss.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,7),wcss)
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
km_best=KMeans(n_clusters=4,init="k-means++",random_state=5)
```

```python
km_best.fit(x_c)
```

```python
y_predicted_km=km_best.predict(x_c)
```

```python
y_predicted_km
```
